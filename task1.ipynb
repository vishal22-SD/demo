# Write a menu driven program to calculate prime,factorial,amstrong,palindrome number

```python
def prime(n):
  if(n<=1):
    return "Not prime"
  else:
    for i in range(2,(n//2+1)):
      if(n%i==0):
        return "Not Prime"
      else:
        return "Prime"
def Factorial(n):
  fact=1
  for i in range(1,n+1):
    fact=fact*i
  return fact
def Palindrome(n):
  val=n
  rev=0
  while(n!=0):
     rem=n%10
     rev=rev*10+rem
     n=n//10
  if(rev==val):
    return "yes palindrome"
  else:
    return "Not palindrome"
def amstrong(n):
  val=n
  length=len(str(n))
  rev=0
  while(n!=0):
     rem=n%10
     rev=rev+(rem**length)
     n=n//10
  if(rev==val):
    return "yes "
  else:
    return "No"


while True:  
    print("\nMAIN MENU")  
    print("1. Check number is Prime or not")  
    print("2. Check number of factorial")  
    print("3. Check number is palindrome or not")  
    print("4. Check number is amstrong or not") 
    print("5. Exit") 
    choice1 = int(input("Enter the Choice:")) 
    if (choice1==1):
      value=int(input())
      print(prime(value))
    if (choice1==2):
      value=int(input())
      print(Factorial(value))
    if (choice1==3):
      value=int(input())
      print(Palindrome(value))
    if (choice1==4):
      value=int(input())
      print(amstrong(value))
    if (choice1==5):
      break
```

# WAP given string is palindrome

```python
def palindrome(n):
   val = ''
   for i in range(len(n) - 1, -1, -1):
      val = val + n[i]
   if (val == n):
      print("palindrome")
   else:
      print("Not palindrome")
str1=input()
palindrome(str1)
```

# WAP menu driven program to sort number using selection sort and insertion sort

```python
def selectionsort(list1):
  for i in range(len(list1)):
     min=i
     for j in range(i+1, len(list1)):
        if list1[min] > list1[j]:
            min = j     
     list1[i], list1[min] = list1[min], list1[i]
  return list1
def insertionsort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j=j-1
        arr[j+1] =key
    return arr
list1 = []
n = int(input("Enter number of elements : "))
for i in range(0, n):
    ele = int(input())
 
    list1.append(ele)
     
while(True):
    print("\nMAIN MENU")  
    print("1. selection sort")  
    print("2. insertion sort")  
    print("3. Exit")   
    choice1 = int(input("Enter the Choice:")) 
    if (choice1==1):
      print(selectionsort(list1))
    if (choice1==2):
      print(insertionsort(list1))
    if (choice1==3):
       break
```

# WAP to calculate largest common subsequence 

```python
def lcs(X, Y, m, n):
  
    if m == 0 or n == 0:
       return 0;
    elif X[m-1] == Y[n-1]:
       return 1 + lcs(X, Y, m-1, n-1);
    else:
       return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n));
str1 =input()
str2 =input()
print ("Length of LCS is ", lcs(str1, str2, len(str1), len(str2)))
```
